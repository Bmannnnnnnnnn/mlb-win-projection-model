```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# =========================
# FILE PATHS
# =========================
hitter_file = "yankees_master_hitter_projections_volpe_pa_reallocated.xlsx"
pitcher_file = "yankees_master_pitcher_projections_conservative_injury_adjusted.xlsx"

# =========================
# MODEL SETTINGS
# =========================
n_simulations = 10000
games = 162
exponent = 2

team_pa_target = 6200
team_ip_target = 1458

league_woba = 0.315
woba_scale = 1.25
league_runs_per_pa = 0.115

runs_scored_uncertainty = 0.07
runs_allowed_uncertainty = 0.08

np.random.seed(42)
# Remove non-2026 Yankees pitchers
pitchers = pitchers[~pitchers["Player"].isin([
    "Weaver, Luke",
    "Williams, Devin"
])].copy()

# =========================
# PLAYER AGES
# Edit these if needed
# =========================
hitter_ages = {
    "Judge, Aaron": 34,
    "Volpe, Anthony": 25,
    "McMahon, Ryan": 31,
    "Bellinger, Cody": 30,
    "Goldschmidt, Paul": 38,
    "Chisholm Jr., Jazz": 28,
    "Stanton, Giancarlo": 36,
    "Dominguez, Jasson": 23,
    "Rice, Ben": 27,
    "Wells, Austin": 26,
    "Grisham, Trent": 29,
    "Rosario, Amed": 30,
    "Cabrera, Oswaldo": 27,
    "Caballero, Jose": 29,
    "Escarra, J.C.": 31,
    "Grichuk, Randal": 34
}

pitcher_ages = {
    "Cole, Gerrit": 35,
    "Rodon, Carlos": 33,
    "Schmidt, Clarke": 30,
    "Fried, Max": 32,
    "Gil, Luis": 28,
    "Williams, Devin": 31,
    "Bednar, David": 31,
    "Weaver, Luke": 32,
    "Doval, Camilo": 28,
    "Cruz, Fernando": 36,
    "Leiter Jr., Mark": 35,
    "Yarbrough, Ryan": 34,
    "Blackburn, Paul": 32,
    "Warren, Will": 27,
    "Schlittler, Cam": 25,
    "Headrick, Brent": 28,
    "Chivilli, Angel": 23,
    "De los Santos, Yerry": 28,
    "Weathers, Ryan": 26
}

# =========================
# AGE ADJUSTMENT FUNCTIONS
# =========================
def hitter_age_factor(age):
    """
    Higher is better for hitters.
    This adjusts offensive contribution, not PA.
    """
    if pd.isna(age):
        return 1.00
    elif age <= 24:
        return 1.03
    elif 25 <= age <= 27:
        return 1.02
    elif 28 <= age <= 30:
        return 1.00
    elif 31 <= age <= 33:
        return 0.97
    elif 34 <= age <= 36:
        return 0.94
    else:
        return 0.90

def pitcher_age_factor(age):
    """
    Higher is worse for pitchers because it multiplies xERA.
    """
    if pd.isna(age):
        return 1.00
    elif age <= 25:
        return 0.98
    elif 26 <= age <= 29:
        return 1.00
    elif 30 <= age <= 32:
        return 1.03
    elif 33 <= age <= 35:
        return 1.07
    else:
        return 1.12

# =========================
# LOAD HITTER DATA
# =========================
hitters_raw = pd.read_excel(hitter_file, sheet_name="Projections", header=None)

header_row = hitters_raw[
    hitters_raw.apply(
        lambda row: row.astype(str).str.contains("Player", case=False).any(),
        axis=1
    )
].index[0]

hitters = pd.read_excel(hitter_file, sheet_name="Projections", header=header_row)

hitters = hitters.rename(columns={
    "Projected PA": "PA",
    "Projected xwOBA": "xwOBA",
    "Projected xSLG": "xSLG"
})

hitters = hitters.dropna(subset=["Player"])
hitters["PA"] = pd.to_numeric(hitters["PA"], errors="coerce")
hitters["xwOBA"] = pd.to_numeric(hitters["xwOBA"], errors="coerce")

hitters = hitters.dropna(subset=["PA", "xwOBA"])

# Add ages
hitters["Age"] = hitters["Player"].map(hitter_ages)
hitters["Age_Factor"] = hitters["Age"].apply(hitter_age_factor)

# Scale team PA to realistic full-season total
pa_scale = team_pa_target / hitters["PA"].sum()
hitters["Scaled_PA"] = hitters["PA"] * pa_scale

# Estimate hitter run contribution using xwOBA
hitters["wRAA_raw"] = ((hitters["xwOBA"] - league_woba) / woba_scale) * hitters["Scaled_PA"]

# Apply age adjustment to offensive contribution
hitters["wRAA_age_adjusted"] = hitters["wRAA_raw"] * hitters["Age_Factor"]

baseline_team_runs = league_runs_per_pa * team_pa_target
projected_runs_scored = baseline_team_runs + hitters["wRAA_age_adjusted"].sum()

print("HITTER AGE ADJUSTMENTS")
print("-" * 50)
print(hitters[["Player", "Age", "Age_Factor", "Scaled_PA", "xwOBA", "wRAA_raw", "wRAA_age_adjusted"]])

# =========================
# LOAD PITCHER DATA
# =========================
pitchers = pd.read_excel(pitcher_file, sheet_name="Projections")

pitchers = pitchers.rename(columns={
    "Projected_IP_Adjusted": "IP",
    "xERA_Proj": "xERA"
})

pitchers = pitchers.dropna(subset=["Player"])
pitchers["IP"] = pd.to_numeric(pitchers["IP"], errors="coerce")
pitchers["xERA"] = pd.to_numeric(pitchers["xERA"], errors="coerce")

pitchers = pitchers.dropna(subset=["IP", "xERA"])

# Add ages
pitchers["Age"] = pitchers["Player"].map(pitcher_ages)
pitchers["Age_Factor"] = pitchers["Age"].apply(pitcher_age_factor)

# Apply age adjustment to xERA
pitchers["xERA_age_adjusted"] = pitchers["xERA"] * pitchers["Age_Factor"]

# Scale team IP to realistic full-season total
ip_scale = team_ip_target / pitchers["IP"].sum()
pitchers["Scaled_IP"] = pitchers["IP"] * ip_scale

# Estimate runs allowed from age-adjusted xERA
pitchers["Projected_Runs_Allowed"] = (
    pitchers["xERA_age_adjusted"] / 9
) * pitchers["Scaled_IP"]

projected_runs_allowed = pitchers["Projected_Runs_Allowed"].sum()

print("\nPITCHER AGE ADJUSTMENTS")
print("-" * 50)
print(pitchers[["Player", "Age", "Age_Factor", "Scaled_IP", "xERA", "xERA_age_adjusted"]])

# =========================
# DYNAMIC INJURY / REPLACEMENT MODEL
# =========================
hitters.loc[hitters["Player"] == "Volpe, Anthony", "PA"] = 500
injury_adjustments = [
    {
        "player": "Gerrit Cole",
        "healthy_ip": 170,
        "projected_ip": 85,
        "replacement_era": 4.85
    },
    {
        "player": "Carlos Rodon",
        "healthy_ip": 160,
        "projected_ip": 120,
        "replacement_era": 4.75
    },
    {
        "player": "Clarke Schmidt",
        "healthy_ip": 130,
        "projected_ip": 25,
        "replacement_era": 4.90
    }
]

extra_runs_allowed = 0

print("\nINJURY / REPLACEMENT ADJUSTMENTS")
print("-" * 50)

for adj in injury_adjustments:
    player = adj["player"]
    row = pitchers[pitchers["Player"] == player]

    if row.empty:
        print(f"{player}: not found in pitcher file")
        continue

    pitcher_xera = row["xERA_age_adjusted"].iloc[0]
    lost_ip = adj["healthy_ip"] - adj["projected_ip"]

    if lost_ip <= 0:
        continue

    extra_runs = ((adj["replacement_era"] - pitcher_xera) / 9) * lost_ip
    extra_runs_allowed += extra_runs

    print(f"{player}")
    print(f"  Healthy IP estimate: {adj['healthy_ip']}")
    print(f"  Adjusted IP estimate: {adj['projected_ip']}")
    print(f"  Lost IP: {lost_ip}")
    print(f"  Age-adjusted pitcher xERA: {pitcher_xera:.2f}")
    print(f"  Replacement ERA: {adj['replacement_era']:.2f}")
    print(f"  Extra runs allowed: {extra_runs:.1f}")
    print()

projected_runs_allowed = projected_runs_allowed + extra_runs_allowed

print(f"Total extra runs allowed from injuries: {extra_runs_allowed:.1f}")
print(f"New projected runs allowed: {projected_runs_allowed:.1f}")

# =========================
# PYTHAGOREAN EXPECTATION
# =========================
def pythagorean_win_pct(rs, ra, exp=2):
    return (rs ** exp) / ((rs ** exp) + (ra ** exp))

base_win_pct = pythagorean_win_pct(
    projected_runs_scored,
    projected_runs_allowed,
    exponent
)

base_projected_wins = base_win_pct * games

print("\nBASE PROJECTION")
print("-" * 50)
print(f"Projected Runs Scored: {projected_runs_scored:.1f}")
print(f"Projected Runs Allowed: {projected_runs_allowed:.1f}")
print(f"Expected Win Percentage: {base_win_pct:.3f}")
print(f"Projected Wins: {base_projected_wins:.1f}")
print(f"Projected Losses: {games - base_projected_wins:.1f}")

# =========================
# MONTE CARLO SIMULATION
# =========================
simulated_runs_scored = np.random.normal(
    projected_runs_scored,
    projected_runs_scored * runs_scored_uncertainty,
    n_simulations
)

simulated_runs_allowed = np.random.normal(
    projected_runs_allowed,
    projected_runs_allowed * runs_allowed_uncertainty,
    n_simulations
)

# Prevent impossible low run totals
simulated_runs_scored = np.maximum(simulated_runs_scored, 300)
simulated_runs_allowed = np.maximum(simulated_runs_allowed, 300)

simulated_win_pct = pythagorean_win_pct(
    simulated_runs_scored,
    simulated_runs_allowed,
    exponent
)

simulated_wins = simulated_win_pct * games

# =========================
# SUMMARY RESULTS
# =========================
print("\nMONTE CARLO RESULTS")
print("-" * 50)
print(f"Average Wins: {np.mean(simulated_wins):.1f}")
print(f"Median Wins: {np.median(simulated_wins):.1f}")
print(f"5th Percentile Wins: {np.percentile(simulated_wins, 5):.1f}")
print(f"95th Percentile Wins: {np.percentile(simulated_wins, 95):.1f}")
print(f"Probability of 90+ Wins: {np.mean(simulated_wins >= 90) * 100:.1f}%")
print(f"Probability of 95+ Wins: {np.mean(simulated_wins >= 95) * 100:.1f}%")
print(f"Probability of 100+ Wins: {np.mean(simulated_wins >= 100) * 100:.1f}%")
print(f"Probability below .500: {np.mean(simulated_wins < 81) * 100:.1f}%")

# =========================
# PLOT RESULTS
# =========================
plt.figure(figsize=(10, 6))
plt.hist(simulated_wins, bins=40)

plt.axvline(np.mean(simulated_wins), linestyle="--", label="Average Wins")
plt.axvline(90, linestyle="--", label="90 Wins")
plt.axvline(100, linestyle="--", label="100 Wins")

plt.title("Monte Carlo Simulation: Yankees Projected Wins with Age Adjustments")
plt.xlabel("Projected Wins")
plt.ylabel("Frequency")
plt.legend()
plt.grid(True)
plt.show()
```
